### Functions for this notebook

```python
def get_aligned_loc_from_index(df,idx_series):
    df_out = df.loc[idx_series.tolist()]
    return df_out

def index_loc_lookup(df,idx_series):
    df_out = dd.map_partitions(get_aligned_loc_from_index,df,idx_series,align_dataframes=False)
    return df_out

def hrm_find_mode(series, max_iter=1000, min_binsize=50):
    working_series = series
    for i in range(max_iter):
        range_max,range_min = np.max(working_series),np.min(working_series)
        midpoint = (range_max+range_min)/2
        above_middle = working_series[working_series>midpoint]
        below_middle = working_series[working_series<=midpoint]
        
        count_above = len(above_middle)
        count_below = len(below_middle)
        
        
        if count_above>count_below:
            working_series = above_middle
        else:
            working_series = below_middle
        
        if i > 0:
            if (len(working_series)<min_binsize) or (last_midpoint==midpoint):
                return np.mean(working_series)

        last_midpoint = midpoint
        
def bootstrap_hrm(series, n_bootstraps=100, max_n_per_bootstrap=100):
    modes = []
    
    series_len = len(series)
    
    n_per_bootstrap = min(series_len,max_n_per_bootstrap)

    for n in range(n_bootstraps):
        modes.append(hrm_find_mode(series.sample(n=n_per_bootstrap)))
    return np.mean(modes)

def get_normal_fovs(fov_series, med_filter_size=5, n_stds=2, max_filter_dev=0.05):
    median_series = sp.ndimage.median_filter(fov_series,size=(med_filter_size,),mode="mirror")
    
    residuals = (fov_series-median_series)/np.mean(median_series)
    gaussian_fit = sp.stats.norm.fit(residuals)
    gaussian_fit = sp.stats.norm(loc=gaussian_fit[0], scale=gaussian_fit[1])

    lower,upper = (-n_stds*gaussian_fit.std(),n_stds*gaussian_fit.std())
    lower,upper = min(-max_filter_dev,lower),max(max_filter_dev,upper)
        
    thr_mask = ((residuals>lower)&(residuals<upper))
    
    return thr_mask
```

### Scaling to uninduced values
 - Comparing measured distribution to the dataset-wide distribution to rescale (measuring both mean and variance)
     - actually got rid of the variance contribution to the scaling, since it is a less reliable estimator
     - especially falls apart when the lineage gets a little heterogenous
 - Could be useful for aligning datasets later
 - Also consider having the derivative values (volume, surface area) recalculated based on the adjustment

```python
nfig = plt.figure(figsize=(30,30))

for i,label in enumerate(values_to_rescale):
    print(i,label)
    trench_means = subsetted_lineage_df_early_trenchid_sorted[label].groupby("Multi-Experiment trenchid").apply(lambda x: np.mean(x),meta=(label,float)).compute()
    global_median_of_means = np.median(trench_means)
    rescaling_lookup = global_median_of_means/trench_means
    rescaling_lookup = rescaling_lookup.to_frame()
    rescaling_lookup.columns = [label + " Rescaling"]
    
    subsetted_lineage_df_trenchid_sorted_label_dd = subsetted_lineage_df_trenchid_sorted[label].to_frame().join(rescaling_lookup)
    rescaled_subsetted_lineage_df_trenchid_sorted = subsetted_lineage_df_trenchid_sorted_label_dd[label]*subsetted_lineage_df_trenchid_sorted_label_dd[label + " Rescaling"]
    rescaled_subsetted_lineage_df_trenchid_sorted.name = label
    
    rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted_label_dd = subsetted_lineage_df_early_subsampled_trenchid_sorted[label].to_frame().join(rescaling_lookup)
    rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted = rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted_label_dd[label]*\
                                                                     rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted_label_dd[label + " Rescaling"]
    rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted.name = label
        
    subsetted_lineage_df_trenchid_sorted["Rescaled: " + label] = rescaled_subsetted_lineage_df_trenchid_sorted
    subsetted_lineage_df_trenchid_sorted[label + " Rescaling"] = subsetted_lineage_df_trenchid_sorted_label_dd[label + " Rescaling"]
    subsetted_lineage_df_early_subsampled_trenchid_sorted["Rescaled: " + label] = rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted
    subsetted_lineage_df_early_subsampled_trenchid_sorted[label + " Rescaling"] = rescaled_subsetted_lineage_df_early_subsampled_trenchid_sorted_label_dd[label + " Rescaling"]
    
    experiment_list = list(subsetted_lineage_df_early_subsampled_trenchid_sorted["Experiment #"].unique())
    
    plt.subplot(4,4,i+1)
    for exp_num in experiment_list:
        exp_subset = subsetted_lineage_df_early_subsampled_trenchid_sorted[subsetted_lineage_df_early_subsampled_trenchid_sorted["Experiment #"] == exp_num]
        plt.hist(exp_subset[label],
             range=(0,np.nanpercentile(subsetted_lineage_df_early_subsampled_trenchid_sorted[label],99)),
             bins=50,alpha=1,histtype='step',label="Experiment " + str(exp_num), density=True)
        
    plt.hist(subsetted_lineage_df_early_subsampled_trenchid_sorted[label],
             range=(0,np.nanpercentile(subsetted_lineage_df_early_subsampled_trenchid_sorted[label],99)),
             bins=50,alpha=1,histtype='step',label="All Experiments", density=True)
    plt.hist(subsetted_lineage_df_early_subsampled_trenchid_sorted["Rescaled: " + label],
             range=(0,np.nanpercentile(subsetted_lineage_df_early_subsampled_trenchid_sorted[label],99)),
             bins=50,alpha=1,histtype='step',label="All Experiments Rescaled", density=True)
    plt.ylabel("# Cells", fontsize=18)
    plt.xlabel(values_to_rescale[i], fontsize=18)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.legend()
                      
plt.savefig("./Trench_correction_replicate_1.png",dpi=500)
plt.show()
```

### Variables over time

```python
target_samples = 100000 
correct_time = True

values_to_rescale_step_2 = ["YFP mean_intensity", 'Nucleoid Mean Intensity Mean']
normalization_fn = [np.max, np.max]
ttl_samples = len(lineage_df_fov_correction)
frac_to_sample = target_samples/ttl_samples
lineage_df_subsample = lineage_df_fov_correction.sample(frac=frac_to_sample).compute()
```
