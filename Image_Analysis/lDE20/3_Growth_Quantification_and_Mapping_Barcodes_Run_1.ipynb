### Functions for this notebook

```python
def filter_df(df,query_list,client=False,repartition=False,persist=False):
    #filter_list must be in df.query format (see pandas docs)
    
    #returns persisted dataframe either in cluster or local
    
    compiled_query = ' and '.join(query_list)
    out_df = df.query(compiled_query)
    if persist:
        if client:
            out_df = client.daskclient.persist(out_df)
        else:
            out_df = out_df.persist()
        
    if repartition:
        init_size = len(df)
        final_size = len(out_df)
        ratio = init_size//final_size
        out_df = out_df.repartition(npartitions=(df.npartitions // ratio) + 1)
        if persist:
            if client:
                out_df = client.daskclient.persist(out_df)
            else:
                out_df = out_df.persist()

    return out_df

def get_first_cell_timepoint(df):
    min_tpts = df.groupby(['Global CellID'])['timepoints'].idxmin().tolist()
    init_cells = df.loc[min_tpts]
    return init_cells

def get_last_cell_timepoint(df):
    max_tpts = df.groupby(['Global CellID'])['timepoints'].idxmax().tolist()
    fin_cells = df.loc[max_tpts]
    return fin_cells

def define_cell_cycle_timepoints_df(query,timepoint_metrics):
    cell_cycle_timepoints_df = query[["timepoints","time (s)","trenchid"] + timepoint_metrics]
    cell_cycle_timepoints_df["Cell Cycle timepoints"] = cell_cycle_timepoints_df.groupby("Global CellID")["timepoints"].apply(lambda x: x - x.iloc[0]).droplevel(1)
    cell_cycle_timepoints_df = cell_cycle_timepoints_df.drop("timepoints", axis=1)
    cell_cycle_timepoints_df = cell_cycle_timepoints_df.rename({"time (s)":"Observation time (s)"}, axis=1)
    cell_cycle_timepoints_df = cell_cycle_timepoints_df.reset_index(drop=False).set_index(["Global CellID","Cell Cycle timepoints"])
    
    return cell_cycle_timepoints_df

def compile_ols_arrs(kymo_df,min_fov,max_fov):
    min_fov_idx = int(f'{int(min_fov):08n}' + '000000000000')
    max_fov_idx = int(f'{int(max_fov):08n}' + '999999999999')
    selected_kymo_df = kymo_df.loc[min_fov_idx:max_fov_idx]
    timepoint_df = selected_kymo_df.groupby(["fov","timepoints"])["time (s)"].first().compute(scheduler='threads')
    timepoint_df = pd.DataFrame(timepoint_df).sort_index()
    timepoint_df["time (s)^2"] = timepoint_df["time (s)"]**2    
    fov_list = timepoint_df.index.get_level_values(0).unique().tolist()
    time_arr = np.array([timepoint_df["time (s)"].loc[f].values for f in fov_list])
    time_arr_accum = np.add.accumulate(time_arr,axis=1)
    time_sq_arr = np.array([timepoint_df["time (s)^2"].loc[f].values for f in fov_list])
    time_sq_arr_accum = np.add.accumulate(time_sq_arr,axis=1)
    return fov_list,time_arr,time_arr_accum,time_sq_arr_accum

def fit_least_squares(log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                     birth_time=None,division_time=None):
    # everything assumes an inclusive time_f
    if time_i>0:
        t_sum = time_arr_accum[fov_idx_in_arr,time_f]-time_arr_accum[fov_idx_in_arr,time_i-1]
        t_sq_sum = time_sq_arr_accum[fov_idx_in_arr,time_f]-time_sq_arr_accum[fov_idx_in_arr,time_i-1]
        
    else:
        t_sum = time_arr_accum[fov_idx_in_arr,time_f]
        t_sq_sum = time_sq_arr_accum[fov_idx_in_arr,time_f]
        
    if (birth_time is not None) and (division_time is not None):
        t_sum = t_sum + birth_time + division_time
        t_sq_sum = t_sq_sum + (birth_time**2) + (division_time**2)
        n_obs = time_f-time_i+3
        time_slice = list(time_arr[fov_idx_in_arr,time_i:time_f+1])
        time_slice = [birth_time] + time_slice + [division_time]
        time_slice = np.array(time_slice)
        
    elif birth_time is not None:
        t_sum = t_sum + birth_time
        t_sq_sum = t_sq_sum + (birth_time**2)
        n_obs = time_f-time_i+2
        time_slice = list(time_arr[fov_idx_in_arr,time_i:time_f+1])
        time_slice = [birth_time] + time_slice
        time_slice = np.array(time_slice)
        
    elif division_time is not None:
        t_sum = t_sum + division_time
        t_sq_sum = t_sq_sum + (division_time**2)
        n_obs = time_f-time_i+2
        time_slice = list(time_arr[fov_idx_in_arr,time_i:time_f+1])
        time_slice = time_slice + [division_time]
        time_slice = np.array(time_slice)
        
    else:
        n_obs = time_f-time_i+1
        time_slice = time_arr[fov_idx_in_arr,time_i:time_f+1]
        
    size_sum = np.sum(log_size_arr)
    time_size_sum = np.sum(log_size_arr*time_slice)
    
    denom = (n_obs*t_sq_sum) - (t_sum**2)
    num = (n_obs*time_size_sum) - (t_sum*size_sum)
    
    beta = num/denom
    y_bar = np.mean(log_size_arr)
    alpha = y_bar-beta*(t_sum/n_obs)
    
    y_hat = alpha + beta*time_slice
    residual_sum_of_squares = np.sum((log_size_arr-y_hat)**2)
    total_sum_of_squares = np.sum((log_size_arr-y_bar)**2)
    r_squared = 1 - (residual_sum_of_squares/total_sum_of_squares)
    if np.isinf(r_squared):
        r_squared = np.NaN
        
    mean_time = np.mean(time_slice)
    
    return beta,alpha,r_squared,mean_time


def fit_least_squares_windowed(timepoint_arr,log_size_arr,fov_idx_in_arr,time_arr,time_arr_accum,time_sq_arr_accum,\
                               birth_time,division_time,ols_window_size=3):
    
    beta_output = []
    r_squared_output = []
    time_output = []
    
    timepoint_arr_len = len(timepoint_arr)
    
    if (birth_time!=None) and (division_time!=None):
        n_timepoints_effective = len(timepoint_arr)+2
        n_windows = n_timepoints_effective-ols_window_size+1
        if n_windows>0:
            window_indices = list(range(n_windows))
            for window_idx in window_indices:
                if window_idx==0:
                    time_i = timepoint_arr[0]
                    working_birth_time = birth_time
                else:
                    time_i = timepoint_arr[window_idx-1]
                    working_birth_time = None
                if window_idx==window_indices[-1]:
                    time_f = timepoint_arr[-1]
                    working_division_time = division_time
                else:
                    time_f = timepoint_arr[window_idx+ols_window_size-2]
                    working_division_time = None
                windowed_log_size_arr = log_size_arr[window_idx:window_idx+ols_window_size]
                beta,_,r_squared,mean_time = fit_least_squares(windowed_log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                birth_time=working_birth_time,division_time=working_division_time)
                beta_output.append(beta)
                r_squared_output.append(r_squared)
                time_output.append(mean_time)
                
    elif division_time!=None:
        n_timepoints_effective = len(timepoint_arr)+1
        n_windows = n_timepoints_effective-ols_window_size+1
        if n_windows>0:
            window_indices = list(range(n_windows))
            for window_idx in window_indices:
                time_i = timepoint_arr[window_idx]
                working_birth_time = None
                if window_idx==window_indices[-1]:
                    time_f = timepoint_arr[-1]
                    working_division_time = division_time
                else:
                    time_f = timepoint_arr[window_idx+ols_window_size-1]
                    working_division_time = None
                
                windowed_log_size_arr = log_size_arr[window_idx:window_idx+ols_window_size]
                beta,_,r_squared,mean_time = fit_least_squares(windowed_log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                birth_time=working_birth_time,division_time=working_division_time)
                beta_output.append(beta)
                r_squared_output.append(r_squared)
                time_output.append(mean_time)
                
    elif birth_time!=None:
        n_timepoints_effective = len(timepoint_arr)+1
        n_windows = n_timepoints_effective-ols_window_size+1
        if n_windows>0:
            window_indices = list(range(n_windows))
            for window_idx in window_indices:
                if window_idx==0:
                    time_i = timepoint_arr[0]
                    working_birth_time = birth_time
                else:
                    time_i = timepoint_arr[window_idx-1]
                    working_birth_time = None
 
                time_f = timepoint_arr[window_idx+ols_window_size-2]
                working_division_time = None
                
                windowed_log_size_arr = log_size_arr[window_idx:window_idx+ols_window_size]
                beta,_,r_squared,mean_time = fit_least_squares(windowed_log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                birth_time=working_birth_time,division_time=working_division_time)
                beta_output.append(beta)
                r_squared_output.append(r_squared)
                time_output.append(mean_time)
    else:
        n_timepoints_effective = len(timepoint_arr)
        n_windows = n_timepoints_effective-ols_window_size+1
        if n_windows>0:
            window_indices = list(range(n_windows))
            for window_idx in window_indices:
                time_i = timepoint_arr[window_idx]
                working_birth_time = None
                time_f = timepoint_arr[window_idx+ols_window_size-1]
                working_division_time = None
                windowed_log_size_arr = log_size_arr[window_idx:window_idx+ols_window_size]
                beta,_,r_squared,mean_time = fit_least_squares(windowed_log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                birth_time=working_birth_time,division_time=working_division_time)
                beta_output.append(beta)
                r_squared_output.append(r_squared)
                time_output.append(mean_time)

    return beta_output,time_output,r_squared_output

def fit_least_squares_windowed_cellgroupby(cell_group,growth_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,ols_window_size=3):
    
    timepoint_arr = cell_group["timepoints"].values
    
    birth_size = cell_group.iloc[0]["Birth: " + growth_metric]
    division_size = cell_group.iloc[0]["Division: " + growth_metric]
    birth_time = cell_group.iloc[0]["Initial time (s)"]
    division_time = cell_group.iloc[0]["Final time (s)"]
    
    if (~np.isnan(birth_size))&(~np.isnan(division_size)):    
        log_size_arr = np.log(np.array([birth_size]+cell_group[growth_metric].values.tolist()+[division_size]))
        
    elif (~np.isnan(birth_size)):
        log_size_arr = np.log(np.array([birth_size]+cell_group[growth_metric].values.tolist()))
        division_time = None
        
    elif (~np.isnan(division_size)):
        log_size_arr = np.log(np.array(cell_group[growth_metric].values.tolist()+[division_size]))
        birth_time = None
    
    else:
        log_size_arr = np.log(cell_group[growth_metric].values)
        birth_time = None
        division_time = None
            
    fov_idx = cell_group["fov"].iloc[0]
    fov_idx_in_arr = np.where(np.array(fov_list)==fov_idx)[0][0]
    
    beta_output,time_output,r_squared_output = fit_least_squares_windowed(timepoint_arr,log_size_arr,fov_idx_in_arr,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                                          birth_time,division_time,ols_window_size=ols_window_size)
    growth_rate_output = np.array([beta_output,time_output,r_squared_output]).T
    growth_rate_output = growth_rate_output.tolist()
    return growth_rate_output

def fit_least_squares_cellcycle(timepoint_arr,log_size_arr,fov_idx_in_arr,time_arr,time_arr_accum,time_sq_arr_accum,birth_time,division_time):
    time_i = timepoint_arr[0]
    time_f = timepoint_arr[-1]
    beta_output,_,r_squared_output,_ = fit_least_squares(log_size_arr,fov_idx_in_arr,time_i,time_f,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                      birth_time=birth_time,division_time=division_time)
    time_output = np.mean(birth_time+division_time)
    
    return beta_output,time_output,r_squared_output

def fit_least_squares_cellcycle_cellgroupby(cell_group,size_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum):
    timepoint_arr = cell_group["timepoints"].values
    # if len(timepoint_arr)>1:
    birth_size = cell_group.iloc[0]["Birth: " + size_metric]
    division_size = cell_group.iloc[0]["Division: " + size_metric]
    birth_time = cell_group.iloc[0]["Initial time (s)"]
    division_time = cell_group.iloc[0]["Final time (s)"]
    log_size_arr = np.log(np.array([birth_size]+cell_group[size_metric].values.tolist()+[division_size]))

    fov_idx = cell_group["fov"].iloc[0]
    fov_idx_in_arr = np.where(np.array(fov_list)==fov_idx)[0][0]
        
    beta_output,time_output,r_squared_output = fit_least_squares_cellcycle(timepoint_arr,log_size_arr,fov_idx_in_arr,time_arr,\
                                                time_arr_accum,time_sq_arr_accum,birth_time,division_time)

    growth_rate_output = [beta_output,time_output,r_squared_output]
    # else:
    #     growth_rate_output = []
    return growth_rate_output

def get_cellcycle_growth_rate_df(query,size_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,\
                                 get_measurement_time=False):
    all_gr = query.groupby(level=0,sort=False).apply(lambda x: fit_least_squares_cellcycle_cellgroupby(x,size_metric,fov_list,time_arr,\
                                                                time_arr_accum,time_sq_arr_accum))
    trenchid_series = query.groupby(level=0,sort=False).apply(lambda x: x["trenchid"].iloc[0])
    
    if get_measurement_time:
        all_gr = pd.concat([all_gr,trenchid_series],axis=1).rename({0:"Growth Measurment",1:"trenchid"},axis=1)
    else:
        all_gr = pd.DataFrame(all_gr, columns=["Growth Measurment"])
        
    all_gr = all_gr.dropna(subset=["Growth Measurment"])
    all_gr = all_gr[all_gr["Growth Measurment"].apply(len)==3]
    
    all_gr["Growth Rate: " + size_metric] = all_gr.apply(lambda x: x["Growth Measurment"][0], axis=1)*3600 #size unit per hr
    all_gr["Growth Rate R2: " + size_metric] = all_gr.apply(lambda x: x["Growth Measurment"][2], axis=1)
    if get_measurement_time:
        all_gr["Measurement time (s)"] = all_gr.apply(lambda x: x["Growth Measurment"][1], axis=1)
    all_gr = all_gr.drop("Growth Measurment", axis=1)
    return all_gr

def get_instantaneous_growth_rate_df(query,growth_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,get_measurement_time=False):
    all_gr = query.groupby(level=0,sort=False).apply(lambda x: fit_least_squares_windowed_cellgroupby(x,growth_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum))
    trenchid_series = query.groupby(level=0,sort=False).apply(lambda x: x["trenchid"].iloc[0])
    
    if get_measurement_time:
        all_gr = pd.concat([all_gr,trenchid_series],axis=1).rename({0:"Growth Measurment",1:"trenchid"},axis=1)
    else:
        all_gr = pd.DataFrame(all_gr, columns=["Growth Measurment"])
    all_gr = all_gr.explode("Growth Measurment").dropna(subset=["Growth Measurment"])
    
    all_gr["Growth Rate Measurement Index"] = all_gr.groupby("Global CellID").cumcount()
    all_gr["Instantaneous Growth Rate: " + growth_metric] = all_gr.apply(lambda x: x["Growth Measurment"][0], axis=1)*3600 #size unit per hr
    all_gr["Instantaneous Growth Rate R2: " + growth_metric] = all_gr.apply(lambda x: x["Growth Measurment"][2], axis=1)
    if get_measurement_time:
        all_gr["Measurement time (s)"] = all_gr.apply(lambda x: x["Growth Measurment"][1], axis=1)
    all_gr = all_gr.drop("Growth Measurment", axis=1)
    return all_gr

def get_growth_and_division_cell_cycle(partition_idx,lineage_cell_cycle_path,lineage_df_path,kymo_df_path,kymo_file_parquet_divisions,trench_score_thr=-75,absolute_time=True,delta_t_min=None,\
                                  size_metrics=["major_axis_length","minor_axis_length","Volume"]): #new version
    
    ## Setting up kymograph and lineage dfs
    lineage_df = dd.read_parquet(lineage_df_path).get_partition(partition_idx).compute(scheduler='threads')
    if len(lineage_df) == 0:
        return (False,partition_idx)
    
    kymo_df = dd.read_parquet(kymo_df_path)
    kymo_df.divisions = kymo_file_parquet_divisions    
    kymo_idx_list = lineage_df["Kymograph FOV Parquet Index"].tolist()
    
    if not absolute_time:
        kymo_df["time (s)"] = kymo_df["timepoints"]*delta_t_min*60.
    
    kymo_time_series = kymo_df["time (s)"].loc[kymo_idx_list].compute(scheduler='threads')
    kymo_time_series.index = lineage_df.index
    lineage_df["time (s)"] = kymo_time_series
            
    reference = filter_df(lineage_df,["`Trench Score` < " + str(trench_score_thr)])
    query = filter_df(lineage_df,["`Trench Score` < " + str(trench_score_thr)])
    
    if len(query) == 0:
        return (False,partition_idx)
    
    init_cells = get_first_cell_timepoint(query).reset_index().set_index("Global CellID").sort_index()
    fin_cells = get_last_cell_timepoint(query).reset_index().set_index("Global CellID").sort_index()

    cell_min_tpt_df = get_first_cell_timepoint(reference).reset_index().set_index("Global CellID").sort_index()
    cell_max_tpt_df = get_last_cell_timepoint(reference).reset_index().set_index("Global CellID").sort_index()
    
    ## adding null entries with NaNs in place of measured values (for cells without mothers/daughters)
    no_cell_entry = cell_max_tpt_df.iloc[0].copy()
    no_cell_entry[size_metrics+["time (s)","timepoints"]] = [np.NaN for i in range(len(size_metrics+["time (s)","timepoints"]))]
    no_cell_entry = pd.DataFrame([no_cell_entry], index=[-1])
    no_cell_entry.index.name = "Global CellID"
    cell_min_tpt_df = pd.concat([no_cell_entry,cell_min_tpt_df], axis=0)
    cell_max_tpt_df = pd.concat([no_cell_entry,cell_max_tpt_df], axis=0)

    ## filter out cells absent from the reference due to some uncharacterize edge case
    in_index_mask = init_cells["Mother CellID"].isin(cell_max_tpt_df.index.unique().tolist())&\
                    init_cells["Sister CellID"].isin(cell_min_tpt_df.index.unique().tolist())&\
                    fin_cells["Daughter CellID 1"].isin(cell_min_tpt_df.index.unique().tolist())&\
                    fin_cells["Daughter CellID 2"].isin(cell_min_tpt_df.index.unique().tolist())
    init_cells = init_cells[in_index_mask]
    fin_cells = fin_cells[in_index_mask]
        
    mother_df = cell_max_tpt_df.loc[init_cells["Mother CellID"].tolist()]
    sister_df = cell_min_tpt_df.loc[init_cells["Sister CellID"].tolist()]
    daughter_1_df = cell_min_tpt_df.loc[fin_cells["Daughter CellID 1"].tolist()]
    daughter_2_df = cell_min_tpt_df.loc[fin_cells["Daughter CellID 2"].tolist()]

    for metric in size_metrics:

        if (metric == "minor_axis_length") or (metric == "Width"):

            init_cells["Birth: " + metric] = init_cells[metric].values
            init_cells["Division: " + metric] = fin_cells[metric].values
            init_cells["Delta: " + metric] = fin_cells[metric].values-init_cells[metric].values

        else:

            interp_mother_final_size = ((init_cells[metric].values + sister_df[metric].values) * mother_df[metric].values)**(1/2)
            sister_frac = init_cells[metric].values/(sister_df[metric].values+init_cells[metric].values)
            init_cells["Birth: " + metric] = (sister_frac*interp_mother_final_size)
            init_cells["Division: " + metric] = (((daughter_1_df[metric].values + daughter_2_df[metric].values) * fin_cells[metric].values)**(1/2))
            init_cells["Delta: " + metric] = init_cells["Division: " + metric].values-init_cells["Birth: " + metric].values
            
            if (metric == "major_axis_length") or (metric == "Length"):
                init_cells["Septum Position"] = daughter_1_df[metric].values # daughter 1 is always towards the old pole
                init_cells["Septum Position Length Normalized"] = daughter_1_df[metric].values/(daughter_1_df[metric].values+daughter_2_df[metric].values)
                init_cells["Septum Displacement"] = np.abs((daughter_1_df[metric].values-daughter_2_df[metric].values)/2)
                init_cells["Septum Displacement Length Normalized"] = np.abs(daughter_1_df[metric].values-daughter_2_df[metric].values)/(2*(daughter_1_df[metric].values+daughter_2_df[metric].values))

    init_cells["Final timepoints"] = daughter_1_df["timepoints"].values #counting a timepoint in which a division occurs as a full timepoint, hacky
    init_cells["Delta Timepoints"] = init_cells["Final timepoints"]-init_cells["timepoints"]

    # if absolute_time:
    interpolated_final_time = (fin_cells["time (s)"].values+daughter_1_df["time (s)"].values)/2 #interpolating under the same assumptions as the size quantification
    interpolated_init_time = ((init_cells["time (s)"].values + mother_df["time (s)"].values)/2)
    init_cells["Final time (s)"] = interpolated_final_time
    init_cells["Initial time (s)"] = interpolated_init_time
    init_cells["Delta time (s)"] = interpolated_final_time-interpolated_init_time
    init_cells["Mid-Cycle time (s)"] = (interpolated_init_time+interpolated_final_time)/2

    query = query.reset_index().set_index(["Global CellID","timepoints"]).sort_index().reset_index(level=1)
    
    delta_t_series = query.groupby(level=0,sort=False)["time (s)"].apply(lambda x: ((x[1:].values - x[:-1].values)))

    init_time_gap = init_cells["time (s)"].values - interpolated_init_time
    final_time_gap = interpolated_final_time - fin_cells["time (s)"].values
    
    init_cells = init_cells.rename(columns={"timepoints":"initial timepoints"})
        
    ### fast computation of the growth rate based on precomputing terms related to shared timepoints
    fov_list = query["fov"].unique().tolist()

    min_fov = fov_list[0]
    max_fov = fov_list[-1]

    fov_list,time_arr,time_arr_accum,time_sq_arr_accum = compile_ols_arrs(kymo_df,min_fov,max_fov)
    
    ## adding interpolated times to query
    query["Initial time (s)"] = init_cells["Initial time (s)"]
    query["Final time (s)"] = init_cells["Final time (s)"]
        
    gr_df_list = []
    for size_idx,size_metric in enumerate(size_metrics): # Havn't decided between mean and median
        ## adding interpolated sizes to query
        query["Birth: " + size_metric] = init_cells["Birth: " + size_metric]
        query["Division: " + size_metric] = init_cells["Division: " + size_metric]
        
        if size_idx == 0:
                        
            gr_df = get_cellcycle_growth_rate_df(query,size_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                 get_measurement_time=True)
        else:
            gr_df = get_cellcycle_growth_rate_df(query,size_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,\
                                                 get_measurement_time=False)

        gr_df_list.append(gr_df)

    gr_df_out = pd.concat(gr_df_list,axis=1)
    gr_df_out = gr_df_out.drop(["trenchid","Measurement time (s)"],axis=1)
    
    init_cells = init_cells.join(gr_df_out)
    init_cells.to_hdf(lineage_cell_cycle_path + "_Temp/temp_output." + str(partition_idx) + ".hdf","data",mode="w",format="table")
    
    return (True,partition_idx)

def get_growth_and_division_cell_cycle_timepoints(partition_idx,timepoints_path,lineage_df_path,kymo_df_path,kymo_file_parquet_divisions,trench_score_thr=-75,absolute_time=True,delta_t_min=None,\
                                  timepoint_metrics=["area","major_axis_length","minor_axis_length","Volume","mCherry mean_intensity"]): #new version
    
    lineage_df = dd.read_parquet(lineage_df_path).get_partition(partition_idx).compute(scheduler='threads')
    if len(lineage_df) == 0:
        return (False,partition_idx)
    ## Setting up kymograph and lineage dfs
    
    kymo_df = dd.read_parquet(kymo_df_path)
    kymo_df.divisions = kymo_file_parquet_divisions
    kymo_idx_list = lineage_df["Kymograph FOV Parquet Index"].tolist()
    
    if not absolute_time:
        kymo_df["time (s)"] = kymo_df["timepoints"]*delta_t_min*60.
    
    kymo_time_series = kymo_df["time (s)"].loc[kymo_idx_list].compute(scheduler='threads')
    kymo_time_series.index = lineage_df.index
    lineage_df["time (s)"] = kymo_time_series
            
    query = filter_df(lineage_df,["`Trench Score` < " + str(trench_score_thr)])
    
    if len(query) == 0:
        return (False,partition_idx)

    query = query.reset_index().set_index(["Global CellID","timepoints"]).sort_index().reset_index(level=1)
    
    cell_cycle_timepoints_df = define_cell_cycle_timepoints_df(query,timepoint_metrics)
    
    cellid_cell_cycle_timepoint_idx = cell_cycle_timepoints_df.reset_index().apply(lambda x: int(f'{int(x["Global CellID"]):016n}{int(x["Cell Cycle timepoints"]):04n}'), axis=1).tolist()
    cell_cycle_timepoints_df["Global CellID-Cell Cycle timepoints"] = cellid_cell_cycle_timepoint_idx
    cell_cycle_timepoints_df = cell_cycle_timepoints_df.reset_index().set_index("Global CellID-Cell Cycle timepoints")
    cell_cycle_timepoints_df.to_hdf(timepoints_path + "_Temp/temp_output." + str(partition_idx) + ".hdf","data",mode="w",format="table")
    
    return (True,partition_idx)
    
def get_growth_and_division_growth_rate(partition_idx,cell_cycle_growth_path,lineage_df_path,kymo_df_path,lineage_cell_cycle_path,\
                                        lineage_cell_cycle_file_parquet_divisions,kymo_file_parquet_divisions,trench_score_thr=-75,\
                                        absolute_time=True,delta_t_min=None,growth_metrics=["major_axis_length","Volume"]): #new version
    
    # partition_i = partition_info['number']
    lineage_df = dd.read_parquet(lineage_df_path).get_partition(partition_idx).compute(scheduler='threads')
    if len(lineage_df) == 0:
        return (False,partition_idx)
    
    kymo_df = dd.read_parquet(kymo_df_path)
    kymo_df.divisions = kymo_file_parquet_divisions
    lineage_cell_cycle_df = dd.read_parquet(lineage_cell_cycle_path).reset_index().set_index("File Parquet Index",sorted=True,divisions=lineage_cell_cycle_file_parquet_divisions)
    lineage_cell_cycle_df = lineage_cell_cycle_df.loc[lineage_df.index[0]:lineage_df.index[-1]].compute(scheduler='threads')
    lineage_cell_cycle_df = lineage_cell_cycle_df.set_index("Global CellID")
    kymo_idx_list = lineage_df["Kymograph FOV Parquet Index"].tolist()

    if not absolute_time:
        kymo_df["time (s)"] = kymo_df["timepoints"]*delta_t_min*60.

    kymo_time_series = kymo_df["time (s)"].loc[kymo_idx_list].compute(scheduler='threads')
    kymo_time_series.index = lineage_df.index
    lineage_df["time (s)"] = kymo_time_series

    #keeping all cells, including those without lineal information
    #given this approach, I need to make sure the initial cell cycle information includes cell cycles
    #that are only recorded at initiation and termination
    #ideally this will be cleaned up after the second export (in the below section)
    query = filter_df(lineage_df,["`Trench Score` < " + str(trench_score_thr)])
    
    ## Adding cell cycle information to the query from seperate calculation
    original_cell_cycle_columns = lineage_cell_cycle_df.columns
    original_query_columns = query.reset_index().columns

    cell_cycle_exclusive_columns = list(set(original_cell_cycle_columns)-set(original_query_columns))
    query = pd.merge(query.reset_index(),lineage_cell_cycle_df[cell_cycle_exclusive_columns],how='left',on='Global CellID').set_index('File Parquet Index')
    
    if len(query) == 0:
        return (False,partition_idx)

    query = query.reset_index().set_index(["Global CellID","timepoints"]).sort_index().reset_index(level=1)
    
    fov_list = query["fov"].unique().tolist()
    min_fov = fov_list[0]
    max_fov = fov_list[-1]

    fov_list,time_arr,time_arr_accum,time_sq_arr_accum = compile_ols_arrs(kymo_df,min_fov,max_fov)
            
    gr_df_list = []
    for growth_idx,growth_metric in enumerate(growth_metrics): # Havn't decided between mean and median
        if growth_idx == 0:
            gr_df = get_instantaneous_growth_rate_df(query,growth_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,get_measurement_time=True)
        else:
            gr_df = get_instantaneous_growth_rate_df(query,growth_metric,fov_list,time_arr,time_arr_accum,time_sq_arr_accum,get_measurement_time=False)

        gr_df["Global CellID-Growth Rate Measurement Index"] = gr_df.reset_index().apply(lambda x: int(f'{int(x["Global CellID"]):016n}{int(x["Growth Rate Measurement Index"]):04n}'), axis=1).tolist()
        gr_df = gr_df.reset_index()
        if growth_idx != 0:
            gr_df = gr_df.drop(["Global CellID","Growth Rate Measurement Index"],axis=1)
        gr_df = gr_df.set_index("Global CellID-Growth Rate Measurement Index")
        gr_df_list.append(gr_df)

        # Initial/Final Growth Rate numbers are probably untrustworthy since noise in delta t from the division 
        # event is probably pretty high. Going to exclude.
    gr_df_out = pd.concat(gr_df_list,axis=1)
    gr_df_out.to_hdf(cell_cycle_growth_path + "_Temp/temp_output." + str(partition_idx) + ".hdf","data",mode="w",format="table")
    
    return (True,partition_idx)

def lineage_partition_mapper(dask_controller,lineage_function,output_path,lineage_df_path,kymo_df_path,lineage_cell_cycle_path="",overwrite=False,**lineage_fn_kwargs): #new version
        
    if overwrite:
        if os.path.exists(output_path + "_progress.pkl"):
            os.remove(output_path + "_progress.pkl")
            os.remove(output_path + "_failed.pkl")
        print("Starting Run.")
    else:
        if os.path.exists(output_path + "_progress.pkl"):
            if os.path.exists(output_path + "_Temp"):
                print("Resuming Run.")
            else:
                print("Run Already Complete.")
                return None
        else:
            print("No Previous Run. Starting Run.")
    
    tr.writedir(output_path + "_Temp",overwrite=overwrite)
    

    # Check for existing files in case of in progress run
    if os.path.exists(output_path + "_progress.pkl"):
        with open(output_path + "_progress.pkl", 'rb') as infile:
            finished_files = pkl.load(infile)
    if os.path.exists(output_path + "_failed.pkl"):
        with open(output_path + "_failed.pkl", 'rb') as infile:
            failed_files = pkl.load(infile)
    else:
        finished_files = []
        failed_files = []
    
    lineage_df = dd.read_parquet(lineage_df_path,calculate_divisions=True)
    kymo_df = dd.read_parquet(kymo_df_path,calculate_divisions=True)
    kymo_file_parquet_divisions = kymo_df.divisions
    
    partition_list = list(range(lineage_df.npartitions))
    dask_controller.futures = {}
    
    if lineage_cell_cycle_path != "":
        lineage_cell_cycle_dd = dd.read_parquet(lineage_cell_cycle_path,calculate_divisions=True)
        lineage_cell_cycle_dd = lineage_cell_cycle_dd.set_index("File Parquet Index",sorted=True)
        lineage_cell_cycle_file_parquet_divisions = lineage_cell_cycle_dd.divisions

        for k,partition_idx in enumerate(partition_list):
            future = dask_controller.daskclient.submit(lineage_function,partition_idx,output_path,lineage_df_path,kymo_df_path,lineage_cell_cycle_path,\
                                                 lineage_cell_cycle_file_parquet_divisions,kymo_file_parquet_divisions,retries=1,**lineage_fn_kwargs)
            # output_futures.append(future)
            dask_controller.futures["Partition: " + str(partition_idx)] = future
    else:
        for k,partition_idx in enumerate(partition_list):
            future = dask_controller.daskclient.submit(lineage_function,partition_idx,output_path,lineage_df_path,kymo_df_path,\
                                                 kymo_file_parquet_divisions,retries=1,**lineage_fn_kwargs)
            # output_futures.append(future)
            dask_controller.futures["Partition: " + str(partition_idx)] = future

    for future in as_completed([val for key,val in dask_controller.futures.items()]):
        result = future.result()
        if result[0]:
            finished_files = finished_files + [result[1]]
        else:
            finished_files = finished_files + [result[1]]
            failed_files = failed_files + [result[1]]
        with open(output_path + "_progress.pkl", 'wb') as outfile:
            pkl.dump(finished_files,outfile)
        with open(output_path + "_failed.pkl", 'wb') as outfile:
            pkl.dump(failed_files,outfile)
        future.cancel()

    unfailed_file_list = sorted(list(set(partition_list) - set(failed_files)))
    temp_output_file_list = [output_path + "_Temp/temp_output." + str(partition_idx) + ".hdf" for partition_idx in unfailed_file_list]
    temp_df = dd.read_hdf(temp_output_file_list,"data",mode="r",sorted_index=True)

    dd.to_parquet(temp_df,output_path,engine='pyarrow',overwrite=True,write_metadata_file=True)
    dask_controller.reset_worker_memory()

    shutil.rmtree(output_path + "_Temp")
    dask_controller.futures = {}

def get_all_growth_and_division_stats(dask_controller,lineage_cell_cycle_path,timepoints_path,cell_cycle_growth_path,\
                                      lineage_df_path,kymo_df_path,trench_score_thr=-75,absolute_time=True,delta_t_min=None,\
                                     size_metrics=["Length","Width","Volume"],timepoint_metrics=["area","Length","Width","Volume","mCherry mean_intensity"],\
                                     growth_metrics=["Length","Volume"],overwrite=False): #new version
    
    lineage_partition_mapper(dask_controller,get_growth_and_division_cell_cycle,lineage_cell_cycle_path,lineage_df_path,\
                         kymo_df_path,overwrite=overwrite,trench_score_thr=trench_score_thr,absolute_time=absolute_time,\
                         delta_t_min=delta_t_min,size_metrics=size_metrics)
    
    lineage_partition_mapper(dask_controller,get_growth_and_division_cell_cycle_timepoints,timepoints_path,lineage_df_path,\
                         kymo_df_path,overwrite=overwrite,trench_score_thr=trench_score_thr,absolute_time=absolute_time,\
                         delta_t_min=delta_t_min,timepoint_metrics=timepoint_metrics)
    
    lineage_partition_mapper(dask_controller,get_growth_and_division_growth_rate,cell_cycle_growth_path,lineage_df_path,\
                         kymo_df_path,lineage_cell_cycle_path=lineage_cell_cycle_path,overwrite=overwrite,trench_score_thr=trench_score_thr,\
                         absolute_time=absolute_time,delta_t_min=delta_t_min,growth_metrics=growth_metrics)

def get_barcode_pheno_kymo_df(phenotype_df, phenotype_kymo_df, barcode_df, trenchid_map, output_kymo_index="FOV Parquet Index"):
    ## phenotype_df_list must contain dfs with trenchids column
    ## can be made more effecient still, with a direct output to parquet
    ## More effecient implementation that splits outputs into two smaller parts (every cell, every trench)
    ## Saves by elmininating some redundent entries
    ## Still filters only for cells that made it to final "phenotype_df" in most cases lineage traced
    
    valid_barcode_df = barcode_df[barcode_df["trenchid"].isin(list(trenchid_map.keys()))].compute()
    barcode_df_mapped_trenchids = valid_barcode_df["trenchid"].apply(lambda x: trenchid_map[x])
    phenotype_df_idx = phenotype_df["trenchid"].unique().compute().tolist()

    valid_init_df_indices = barcode_df_mapped_trenchids.isin(phenotype_df_idx)
    barcode_df_mapped_trenchids = barcode_df_mapped_trenchids[valid_init_df_indices]
    barcode_df_mapped_trenchids_list = barcode_df_mapped_trenchids.tolist()
    final_valid_barcode_df_indices = barcode_df_mapped_trenchids.index.to_list()

    called_df = barcode_df.loc[final_valid_barcode_df_indices]
    called_df["phenotype trenchid"] = barcode_df_mapped_trenchids
    called_df["phenotype trenchid"] = called_df["phenotype trenchid"].astype(int)
    called_df = called_df.drop(["Barcode Signal"], axis=1)
    called_df = called_df.reset_index().set_index("phenotype trenchid",drop=True,sorted=False)

    output_phenotype_kmyo_df = phenotype_kymo_df.rename(columns={"trenchid":"phenotype trenchid"})
    output_phenotype_kmyo_df = output_phenotype_kmyo_df.reset_index().set_index("phenotype trenchid",drop=True,sorted=True)
    output_phenotype_kmyo_df = output_phenotype_kmyo_df.loc[barcode_df_mapped_trenchids_list]
    called_df = called_df.repartition(divisions = output_phenotype_kmyo_df.divisions).persist()
    wait(called_df)
    output_phenotype_kmyo_df = output_phenotype_kmyo_df.merge(called_df,how='inner',left_index=True,right_index=True)
    output_phenotype_kmyo_df = output_phenotype_kmyo_df.reset_index().set_index(output_kymo_index)

    return output_phenotype_kmyo_df,barcode_df_mapped_trenchids

def get_output_phenotype_df(phenotype_df,barcode_df_mapped_trenchids,output_index="File Parquet Index"):
    
    phenotype_df_idx = phenotype_df["trenchid"].unique().compute().tolist()
    valid_init_df_indices = barcode_df_mapped_trenchids.isin(phenotype_df_idx)
    barcode_df_mapped_trenchids = barcode_df_mapped_trenchids[valid_init_df_indices]
    barcode_df_mapped_trenchids_list = barcode_df_mapped_trenchids.tolist()
    
    output_phenotype_df = phenotype_df.rename(columns={"trenchid":"phenotype trenchid"})
    output_phenotype_df = output_phenotype_df.reset_index().set_index("phenotype trenchid",drop=True,sorted=True)
    
    output_phenotype_df = output_phenotype_df.loc[barcode_df_mapped_trenchids_list]
    output_phenotype_df = output_phenotype_df.reset_index().set_index(output_index) 
    
    return output_phenotype_df
```

#### Remove Single Timepoints
- Single timepoints appear due to segmentation error
- Max possible observational period with a single reported value is 20 mins, so this is sound from a biological perspective (no divisions possible under 20 mins)
- Note the difference between the instantaneous and cell cycle dataframes arises from the lineal filtering step on the cell cycle dataframe
    - i.e. the high number of singletons likely arises from cells without lineal relations

```python
def filter_instantaneous_growth_timepoints(instantaneous_growth_df_output,min_timepoint_thr=2,ols_window=3):
    global_cellids_past_thr = instantaneous_growth_df_output.groupby("Global CellID")["Growth Rate Measurement Index"].max()>=(min_timepoint_thr-ols_window+2)
    global_cellids_past_thr_list = global_cellids_past_thr[global_cellids_past_thr].index.tolist()
    instantaneous_growth_df_output = instantaneous_growth_df_output[instantaneous_growth_df_output["Global CellID"].isin(global_cellids_past_thr_list)]
    return instantaneous_growth_df_output
def filter_instantaneous_growth_timepoints_exact(instantaneous_growth_df_output,timepoint_thr=2,ols_window=3):
    global_cellids_past_thr = instantaneous_growth_df_output.groupby("Global CellID")["Growth Rate Measurement Index"].max()==(timepoint_thr-ols_window+2)
    global_cellids_past_thr_list = global_cellids_past_thr[global_cellids_past_thr].index.tolist()
    instantaneous_growth_df_output = instantaneous_growth_df_output[instantaneous_growth_df_output["Global CellID"].isin(global_cellids_past_thr_list)]
    return instantaneous_growth_df_output
```
