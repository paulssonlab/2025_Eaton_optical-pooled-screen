```python
def parallel_norm_soft_dtw(X,chunk_size=200):
    X_dask = da.from_array(X,chunks=(chunk_size,X.shape[1],X.shape[2]))
    soft_dtw_arr = da.blockwise(cdist_soft_dtw,"ik",X_dask,"itd",X_dask,"ktd",concatenate=True).compute()
    d_ii = np.diag(soft_dtw_arr)
    norm_soft_dtw_arr = soft_dtw_arr - (.5 * (d_ii.reshape((-1, 1)) + d_ii.reshape((1, -1))))
    return norm_soft_dtw_arr

def norm_soft_dtw(X):
    soft_dtw_arr = cdist_soft_dtw(X)
    d_ii = np.diag(soft_dtw_arr)
    norm_soft_dtw_arr = soft_dtw_arr - (.5 * (d_ii.reshape((-1, 1)) + d_ii.reshape((1, -1))))
    return norm_soft_dtw_arr

def inter_exp_distances(x_chunk,exp_ids=[1,2]):
    exp_chunk_arr = []
    mean_exp_dist_dict = {}
    
    exp_chunk_lens = [len(x_chunk[x_chunk['Experiment #']==i]) for i in exp_ids]
    
    if np.all([chunk_len>1 for chunk_len in exp_chunk_lens]):

        for i in exp_ids:
            exp_chunk_arr.append(np.swapaxes(np.stack(x_chunk[x_chunk['Experiment #']==i]['Feature Vector'].values),1,2))

        exp_chunk_lens = [len(exp_chunk) for exp_chunk in exp_chunk_arr]
        exp_chunk_intervals = np.concatenate([np.array([0]),np.add.accumulate(exp_chunk_lens)])

        if np.all([chunk_len>1 for chunk_len in exp_chunk_lens]):
            exp_chunk_arr = np.concatenate(exp_chunk_arr)

            dist_out = norm_soft_dtw(exp_chunk_arr)

            for i in range(len(exp_ids)):
                for j in range(i,len(exp_ids)):
                    dist_section = dist_out[exp_chunk_intervals[i]:exp_chunk_intervals[i+1],exp_chunk_intervals[j]:exp_chunk_intervals[j+1]]
                    if i == j:
                        dist_section = np.trim_zeros(np.sort(np.triu(dist_section).flatten()))
                    else:
                        dist_section = dist_section.flatten()
                    mean_exp_dist_dict[(exp_ids[i],exp_ids[j])] = np.mean(dist_section)

    else:
        for i in range(len(exp_ids)):
            for j in range(i,len(exp_ids)):
                mean_exp_dist_dict[(exp_ids[i],exp_ids[j])] = np.NaN
    
    return mean_exp_dist_dict


def search_go(ns2assoc, obodag, inv_gene_to_id, go_term):
    namespace_abbv = {"biological_process":"BP","molecular_function":"MF","cellular_component":"CC"}
    
    print("Searching for " + str(obodag[go_term].name))
    namespace = namespace_abbv[obodag[go_term].namespace]
    child_goterms = list(obodag[go_term].get_all_children())
    gene_list = [inv_gene_to_id[key] for key,val in ns2assoc[namespace].items() if go_term in val]
    for child_goterm in child_goterms:    
        gene_list += [inv_gene_to_id[key] for key,val in ns2assoc[namespace].items() if child_goterm in val]
    gene_list = sorted(list(set(gene_list)))
    return gene_list

def selection_fn(item, gene_name):
    is_gene = item["Gene"]==gene_name
    if is_gene:
        return item["TargetID"]
    else:
        return 0

def highlight_gene_group(an_df,selection_list):

    highlight_genes_df = copy.deepcopy(an_df)
    
    selection_list = sorted(list(set(highlight_genes_df.obs["Gene"].unique().tolist())&set(selection_list)))

    for i,selected_gene in enumerate(selection_list):
        selected_series = (highlight_genes_df.obs["Gene"]==selected_gene).astype("category")
        selected_series = selected_series.cat.reorder_categories([True,False])
        highlight_genes_df.obs["Selected Genes: " + str(i)] = selected_series
            
    selected_series = (highlight_genes_df.obs["Gene"].isin(selection_list)).astype("category")
    selected_series = selected_series.cat.reorder_categories([True,False])
    highlight_genes_df.obs["All Genes"] = selected_series

    # selected_series = (paga_df.obs["Gene"]=="ftsZ").astype(float)
    # selected_series[selected_series==0.] = np.NaN
    # paga_df.obs["Selected Genes"] = selected_series

    fig = sc.pl.umap(highlight_genes_df, title=selection_list + ["All Genes"], color=["Selected Genes: " + str(i) for i in range(len(selection_list))] + ["All Genes"],
                     groups=[True], show=False, legend_loc='right margin', add_outline=False, size=50, return_fig=True,
                     palette = {True:"red",False:"lightgrey"})# palette ={}

    return fig

def highlight_sgrnas(an_df,selection_list):

    highlight_genes_df = copy.deepcopy(an_df)
    highlight_genes_df.obs["tempindex"] = highlight_genes_df.obs.index
    
    selection_list = sorted(list(set(highlight_genes_df.obs["tempindex"].unique().tolist())&set(selection_list)))        
    selected_series = (highlight_genes_df.obs["tempindex"].isin(selection_list)).astype("category")
    selected_series = selected_series.cat.reorder_categories([True,False])
    highlight_genes_df.obs["All sgRNAs"] = selected_series

    # selected_series = (paga_df.obs["Gene"]=="ftsZ").astype(float)
    # selected_series[selected_series==0.] = np.NaN
    # paga_df.obs["Selected Genes"] = selected_series

    fig = sc.pl.umap(highlight_genes_df, title="All sgRNAs", color="All sgRNAs",
                     groups=[True], show=False, legend_loc='right margin', add_outline=False, size=50, return_fig=True,
                     palette = {True:"red",False:"lightgrey"})# palette ={}

    return fig

def get_top_genes(x,n_top=2,n_gene_names_per_line=2):
    genes,counts = np.unique(x,return_counts=True)
    sorted_genes = genes[np.argsort(counts)[::-1]]
    sorted_genes = list(sorted_genes[:n_top])
    sorted_genes = [".\n".join(gene.split()[:n_gene_names_per_line]) for gene in sorted_genes]
    return sorted_genes

def get_recall_df(an_df,minimum_total_count=2,recall_metric="TargetID",cluster_label="leiden"):
    an_df_copy = copy.deepcopy(an_df)
    recall_metric_count_dict = dict(zip(*np.unique(an_df_copy.obs[recall_metric],return_counts=True)))
    all_counts = np.sum(list(recall_metric_count_dict.values()))
    
    top_2_genes_series = an_df_copy.obs.groupby(cluster_label).apply(lambda x: "\n".join(get_top_genes(x["Gene"])))
    top_2_genes_df = pd.DataFrame(top_2_genes_series).rename(columns={0:"Top_2_Genes"})
    top_2_genes_df_columns = top_2_genes_df.columns
    an_df_addition = an_df_copy.obs[[cluster_label]].reset_index().merge(top_2_genes_df, on=cluster_label).set_index('oDEPool7_id')
    an_df_copy.obs[top_2_genes_df_columns] = an_df_addition[top_2_genes_df_columns]
    
    recall_count_in_cluster = an_df_copy.obs.groupby(cluster_label).apply(lambda x: dict(zip(*np.unique(x[recall_metric],return_counts=True))))
    recall_count_all_clusters = an_df_copy.obs.groupby(cluster_label).apply(lambda x: {key:recall_metric_count_dict[key] for key in np.unique(x[recall_metric])})
    recall_df = pd.concat([recall_count_in_cluster,recall_count_all_clusters],axis=1).rename(columns={0:"In Cluster",1:"All Clusters"})

    recall_df[recall_metric + "_Recall_List_" + cluster_label] = recall_df.apply(lambda x: [x["In Cluster"][key]/x["All Clusters"][key] for key in x["In Cluster"].keys() if x["All Clusters"][key]>=minimum_total_count], axis=1)
    recall_df = recall_df[recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(len)>0]
    
    recall_df[recall_metric + "_Mean_Recall_" + cluster_label] = recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(lambda x: np.mean(x))
    recall_df[recall_metric + "_Median_Recall_" + cluster_label] = recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(lambda x: np.median(x))
    recall_df[recall_metric + "_Max_Recall_" + cluster_label] = recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(lambda x: np.max(x))
    recall_df[recall_metric + "_Min_Recall_" + cluster_label] = recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(lambda x: np.min(x))
    recall_df[recall_metric + "_Top2_Recall_" + cluster_label] = recall_df[recall_metric + "_Recall_List_" + cluster_label].apply(lambda x: sorted(x)[-2] if len(x)>1 else 0.)
    recall_df = recall_df.drop(columns=["In Cluster","All Clusters"])
    
    return recall_df

def get_recall_from_intersection(agg_intersection_flat,cluster_str,jackknife_ttl=len(jackknife_indices)):
    sgrnaids,counts = np.unique(agg_intersection_flat[cluster_str],return_counts=True)
    recall_vals = counts/jackknife_ttl
    recall_dict = {sgrnaid:recall_vals[i] for i,sgrnaid in enumerate(sgrnaids)}
    return recall_dict
```

### Jackknife Cluster Stability

```python
for resolution_idx in range(len(resolution_list)):
    cluster_level = "L" + str(resolution_idx)
    
    ARI_list = []
    jaccard_list = []
    recall_list = []
    intersection_list = []
    for jackknife_i in jackknife_indices:
        jackknife_clusters = an_df_merged_jackknife_list[jackknife_i].obs[cluster_level]
        jackknife_sgrnaids = jackknife_clusters.index.unique().astype(int).tolist()
        true_sgrnaids = an_df_clustree.obs[cluster_level].index.unique().astype(int).tolist()
        shared_sgrnaids = sorted(list(set(jackknife_sgrnaids)&set(true_sgrnaids)))
        shared_sgrnaids = [str(item) for item in shared_sgrnaids]
        
        clusters_i = an_df_clustree.obs[cluster_level].loc[shared_sgrnaids]
        clusters_j = jackknife_clusters.loc[shared_sgrnaids]
    
        ARI = skl.metrics.adjusted_rand_score(clusters_i,clusters_j)
        ARI_list.append(ARI) 
    
        main_cluster_list = clusters_i.unique()
        jaccard_score_dict = {}
        recall_score_dict = {}
        intersection_sgrna_dict = {}
        for main_cluster in main_cluster_list:
            clusters_i_single_cluster = (clusters_i==main_cluster)
            clusters_j_single_cluster_list = clusters_j[clusters_i_single_cluster].unique()
    
            jaccard_score_list = []
            recall_score_list = []
            intersection_sgrna_list = []
            for clusters_j_single_cluster in clusters_j_single_cluster_list:
                n_union = np.sum((clusters_j==clusters_j_single_cluster)|clusters_i_single_cluster)
                intersection_df = (clusters_j==clusters_j_single_cluster)&clusters_i_single_cluster
                intersection_sgrnaids = intersection_df[intersection_df].index.tolist()
                n_intersection = np.sum(intersection_df)
                jaccard_score = n_intersection/n_union
                recall_score = n_intersection/np.sum(clusters_i_single_cluster)
                
                jaccard_score_list.append(jaccard_score)
                recall_score_list.append(recall_score)
                intersection_sgrna_list.append(intersection_sgrnaids)
                
            jaccard_score_dict[main_cluster] = np.max(jaccard_score_list)
            recall_score_dict[main_cluster] = recall_score_list[np.argmax(jaccard_score_list)]
            intersection_sgrna_dict[main_cluster] = intersection_sgrna_list[np.argmax(jaccard_score_list)]
        
        jaccard_list.append(jaccard_score_dict)
        recall_list.append(recall_score_dict)
        intersection_list.append(intersection_sgrna_dict)
        
    mean_jaccard_score = {key:np.mean([item[key] for item in jaccard_list]) for key in jaccard_list[0].keys()}
    mean_jaccard_score = pd.Series(mean_jaccard_score)
    mean_jaccard_score.name=cluster_level + " Jackknife Jaccard"

    mean_recall_score = {key:np.mean([item[key] for item in recall_list]) for key in recall_list[0].keys()}
    mean_recall_score = pd.Series(mean_recall_score)
    mean_recall_score.name=cluster_level + " Jackknife Recall"
    
    agg_intersection = {key:[item[key] for item in intersection_list] for key in intersection_list[0].keys()}
    agg_intersection_flat= {key:[val for sublist in nested_list for val in sublist] for key,nested_list in agg_intersection.items()}
    
    recall_output_df = []
    for main_cluster in main_cluster_list:
        true_sgrnaids = an_df_clustree.obs[cluster_level].index.unique().tolist()
        recall_dict = get_recall_from_intersection(agg_intersection_flat,main_cluster)
        recall_output = {sgrnaid:recall_dict[sgrnaid] for sgrnaid in true_sgrnaids if sgrnaid in recall_dict.keys()}
        recall_output = pd.Series(recall_output)
        recall_output.name=cluster_level + " Jackknife sgRNA Recall"
        recall_output_df.append(recall_output)
    recall_output_df = pd.concat(recall_output_df, axis=0)
    
    an_df_clustree.obs = an_df_clustree.obs.join(mean_jaccard_score,on=cluster_level)
    an_df_clustree.obs = an_df_clustree.obs.join(mean_recall_score,on=cluster_level)
    an_df_clustree.obs = an_df_clustree.obs.join(recall_output_df,on="oDEPool7_id")
```
