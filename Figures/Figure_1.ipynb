## Figure 1B

```python
chip_view = tr.whole_chip_viewer("/home/de64/scratch/de64/sync_folder/2023-01-14_lDE20_Run_9/Growth_Division",thumbnail=False)
```

### Full Scale

```python
time_idx = 0
channel_idx = 0
resize_scale = 0.01

img = chip_view.main_array[channel_idx,time_idx]
rescaled_img = da.map_blocks(sk.transform.rescale,img,resize_scale,preserve_range=True,anti_aliasing=True).astype("uint16")
rescaled_img = rescaled_img.compute()
```

```python
fig = plt.figure(figsize=(rescaled_img.shape[0]//50,rescaled_img.shape[1]//50))
plt.axis('off')
plt.imshow(rescaled_img,cmap="Greys_r",vmin=0,vmax=3000)
plt.savefig("/home/de64/group/de64/CRISPRi_Libraries/dev_notebooks/2024-11-23_Figure_Notebooks/Figure_1/full_scale.png",dpi=500,bbox_inches="tight")
```

### 5x Scale

```python
resize_scale = 0.05*5

img = chip_view.main_array[channel_idx,time_idx]
rescaled_img = da.map_blocks(sk.transform.rescale,img,resize_scale,preserve_range=True,anti_aliasing=True).astype("uint16")
rescaled_img = rescaled_img.compute()

fig = plt.figure(figsize=(20,20))
plt.imshow(rescaled_img[:3000,-3000:],cmap="Greys_r",vmin=0,vmax=5000)
plt.axis('off')
plt.savefig("/home/de64/group/de64/CRISPRi_Libraries/dev_notebooks/2024-11-23_Figure_Notebooks/Figure_1/5x_scale.png",dpi=500,bbox_inches="tight")
```

### Kymograph

```python
##mirroring previous
kyimg = chip_view.main_array[0,:,:2960,-2960:].compute()
kyimg = kyimg[:,650:1200,200:1000]

kyhand = tr.kymo_handle()
kyimg_crop = kyimg[:20,310:470,644:663]
kyhand.import_wrap(kyimg_crop)
kyhand = kyhand.return_unwrap()

plt.imshow(kyhand,cmap="Greys_r",vmin=0,vmax=5000)
plt.vlines([(kyimg_crop.shape[2]*t) for t in range(1,kyimg_crop.shape[0])],-0.5,160,linestyle="-",color="white",linewidth=1.5,zorder=20)
plt.axis('off')
plt.savefig("/home/de64/group/de64/CRISPRi_Libraries/dev_notebooks/2024-11-23_Figure_Notebooks/Figure_1/kymograph.png",dpi=500,bbox_inches="tight")
```
