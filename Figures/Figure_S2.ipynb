```python
def get_magenta_green_overlay(green_arr,mag_arr,green_arr_min,green_arr_max,mag_arr_min,\
                              mag_arr_max,mag_weight=1.,green_weight=1.):
    
    # Expand dimensions to make them 3D arrays
    image1 = np.expand_dims(green_arr, axis=-1)
    image2 = np.expand_dims(mag_arr, axis=-1)
    
    # Make image1 green
    img_green = np.zeros((image1.shape[0], image1.shape[1], 3), dtype=np.uint16)
    img_green[:, :, 1] = image1[:,:,0]
    
    # Make image2 magenta
    img_magenta = np.zeros((image2.shape[0], image2.shape[1], 3), dtype=np.uint16)
    img_magenta[:, :, 0] = image2[:,:,0]
    img_magenta[:, :, 2] = image2[:,:,0]
    
    # # Apply contrast to the green channel
    img_green[:, :, 1] = sk.exposure.rescale_intensity(np.clip(img_green[:, :, 1],green_arr_min,green_arr_max)).astype('uint16')
    
    # Apply contrast to the magenta channel
    img_magenta[:, :, 0] = sk.exposure.rescale_intensity(np.clip(img_magenta[:, :, 0],mag_arr_min,mag_arr_max)).astype('uint16')
    img_magenta[:, :, 2] = sk.exposure.rescale_intensity(np.clip(img_magenta[:, :, 2],mag_arr_min,mag_arr_max)).astype('uint16')
    
    merged_image = ((img_green*green_weight)+(img_magenta*mag_weight)).astype('uint16')
    return merged_image
```

```python
## Setting up scale bar
scale_bar_size_um = 6

scalebar_color = 'white'  # Color of the scale bar
# scalebar_y_position = 230  # Y-position of the scale bar in data units
scalebar_y_position = 185  # Y-position of the scale bar in data units
# scalebar_x_position = 200  # X-position of the scale bar in data units
scalebar_x_position = 322  # X-position of the scale bar in data units

pixel_microns = tr.pandas_hdf5_handler("/home/de64/scratch/de64/sync_folder/2021-10-21_lDE15_Final_1/GFP/metadata.hdf5").\
read_df("global",read_metadata=True).metadata['pixel_microns']
scale_bar_size_pixels = scale_bar_size_um//pixel_microns
kymo_df = dd.read_parquet("/home/de64/scratch/de64/sync_folder/2021-10-21_lDE15_Final_1/GFP/kymograph/metadata",calculate_divisions=True)
sub_kymo_df = kymo_df.sample(frac=0.01).compute()
```

```python
n_trenches = 20

#fetch trenches
selected_trenches = sub_kymo_df.sample(n=n_trenches)

gfp_output = []
rfp_output = []
seg_output = []
for sub_idx,selected_trench in enumerate(selected_trenches.iterrows()):
    _,selected_trench = selected_trench
        
    file_idx,trench_idx,orient = selected_trench['File Index'],selected_trench['File Trench Index'],\
    selected_trench["lane orientation"]
    
    with h5py.File("/home/de64/scratch/de64/sync_folder/2021-10-21_lDE15_Final_1/GFP/kymograph/kymograph_" +\
                   str(file_idx) + ".hdf5", "r") as infile:
        rfp_data = infile["RFP-Penta"][trench_idx,-1,:,1:-1]
        gfp_data = infile["GFP-Penta"][trench_idx,-1,:,1:-1]
    with h5py.File("/home/de64/scratch/de64/sync_folder/2021-10-21_lDE15_Final_1/GFP/fluorsegmentation/segmentation_" +\
                   str(file_idx) + ".hdf5", "r") as infile:
        seg_data = infile["data"][trench_idx,-1,:,1:-1]
    if orient == "top":
        rfp_data = rfp_data[::-1]
        gfp_data = gfp_data[::-1]
        seg_data = seg_data[::-1]
    
    gfp_output.append(gfp_data)
    rfp_output.append(rfp_data)

    seg_data[seg_data>0] = seg_data[seg_data>0]+(1000*sub_idx)
    seg_output.append(seg_data)


gfp_output = np.concatenate(gfp_output,axis=1)
rfp_output = np.concatenate(rfp_output,axis=1)
seg_output = np.concatenate(seg_output,axis=1)
seg_output, _, _ = sk.segmentation.relabel_sequential(seg_output)

seg_borders_list = []
for i in range(1,np.max(seg_output)+1):
    seg_borders = sk.measure.find_contours(sk.filters.gaussian(seg_output == i,sigma=2), level=0.5)[0]
    seg_borders_list.append(seg_borders)
```

## Figure S2E

```python
def get_confusion_mat(df):
    TP = np.sum((df["dark_gfp"] == False) & (df["Measured Dark GFP"] == False))
    TN = np.sum((df["dark_gfp"] == True) & (df["Measured Dark GFP"] == True))
    FP = np.sum((df["dark_gfp"] == False) & (df["Measured Dark GFP"] == True))
    FN = np.sum((df["dark_gfp"] == True) & (df["Measured Dark GFP"] == False))

    error = (FP + FN) / (TP + TN + FP + FN)
    FP_error = FP / (TP + TN + FP + FN)
    FN_error = FN / (TP + TN + FP + FN)

    return error, FP_error, FN_error
```

```python
hamming_filters = list(range(1, 5))
hamming_n_barcodes = []
hamming_errors = []
unrounded_hamming_errors = []
for i in hamming_filters:
    filtered_df = output_single_trenchid_df[
        output_single_trenchid_df["Closest Hamming Distance"] >= i
    ]
    n_barcode = len(filtered_df)
    error, FP_error, FN_error = get_confusion_mat(filtered_df)
    unrounded_hamming_errors.append(error*100)
    error = np.round(100 * error, decimals=2)
    hamming_errors.append(error)
    hamming_n_barcodes.append(n_barcode)

fig = plt.figure(figsize=(8, 6))

sns.lineplot(x=hamming_filters, y=hamming_errors, linewidth=4, marker="o", markersize=15,zorder=1)
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: int(x)))
plt.xlabel("Minimum Hamming Distance", fontsize=26)
plt.ylabel(r"$Err_{Tot}$ (%)", fontsize=26)
plt.xticks([1, 2, 3, 4], fontsize=26)
plt.yticks(fontsize=26,)
plt.hlines(hamming_errors[0],hamming_filters[0],hamming_filters[-1]+0.05,color="black",\
           linestyle="--",zorder=-1,linewidth=3)
plt.ylim(0.0, 1.2)

# Get current axes
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(5)
ax.tick_params(axis='both', width=5, length=10)  # For major ticks
sns.despine()

sns.despine()
plt.savefig("/home/de64/group/de64/CRISPRi_Libraries/dev_notebooks/2024-11-23_Figure_Notebooks/S2/Hamming_Dist_vs_Error.png", dpi=300, bbox_inches="tight")
plt.show()
```
